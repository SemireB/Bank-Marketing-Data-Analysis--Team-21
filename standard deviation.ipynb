Determine standard deviation, range, IQR for numerical varibales + visual summary of the output

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset 
file_path = '/Users/72xs/team project - bank marketing /bank+marketing/bank/bank-full.csv'
data = pd.read_csv(file_path, sep=';')

# Select numerical columns
numerical_columns = ['age', 'balance', 'day', 'duration', 'campaign', 'pdays', 'previous']
numerical_data = data[numerical_columns]

# Calculate statistics
stats_summary = numerical_data.describe().T  # Basic descriptive stats (mean, std, min, max, etc.)
stats_summary['std_dev'] = numerical_data.std()  # Standard deviation
stats_summary['range'] = stats_summary['max'] - stats_summary['min']  # Range
stats_summary['IQR'] = stats_summary['75%'] - stats_summary['25%']  # Interquartile range

# Display summary statistics
print(stats_summary[['mean', 'std_dev', 'range', 'IQR']])

# Generate visual summaries
plt.figure(figsize=(15, 10))  # Set figure size for multiple plots

# Create a boxplot for each numerical variable
for i, column in enumerate(numerical_columns, 1):
    plt.subplot(3, 3, i)  # Organize plots in a grid
    sns.boxplot(x=numerical_data[column], color='skyblue')  # Boxplot
    plt.title(column)  # Add title to each subplot

plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()  # Display the plots
```
